```python
#statics 
stats = np.array([[1,2,3],[4,5,6]])
a = np.min(stats)
b = np.max(stats)
c = np.sum(stats)
d = np.min(stats,axis=1)
e = np.max(stats,axis=0)
f = np.sum(stats,axis=0)
print(a)
print(b)    
print(c)
print(d)
print(e)
print(f)
```

```python
#reorganizing arrays
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)
after = before.reshape((2,2,2))
print(after)
```

```python
#vertically stacking vectors
v1 = np.array([1,2,3,4]) 
v2 = np.array([5,6,7,8,])
np.vstack((v1,v2,v2,v2,v1))
```

```python
#horizontal stacking
v1 = np.array([1,2,3,4]) 
v2 = np.array([5,6,7,8,])
a = np.hstack((v1,v2,v2,v2,v1))
#a.reshape(5,4)
print(a)    
```

```python
#Miscellaneous
#Load Data from File
filedata = np.loadtxt('file.txt', delimiter=',')
#filedata = np.genfromtxt('file.txt', delimiter=',')
filedata = filedata.astype('int32')
print(filedata)
```

```python
#boolean masking and advanced indexing
filedata > 50
```

```python
filedata[filedata > 50]
```

```python
a = filedata[(filedata > 50) & (filedata < 100)]
print(a)
b = (((filedata > 50) & (filedata < 100)))
print(b)
c = (~((filedata > 50) & (filedata < 100))) 
print(c,end = "")
d = filedata[filedata == 0]
print(d)
```
